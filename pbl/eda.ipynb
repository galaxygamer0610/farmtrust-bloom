# Exploratory Data Analysis: AgriRiskFin Dataset
This notebook analyzes the **Pre-Cleaned Raw Data** vs the **Post-Cleaned Data** to highlight the anomalies caught and fixed by the data engineering pipeline.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style="whitegrid", palette="muted")
```

```python
# Load Datasets
df_raw = pd.read_csv('AgriRiskFin_Dataset.csv')
df_clean = pd.read_csv('AgriRiskFin_Dataset_Cleaned.csv')

print(f"Raw Shape: {df_raw.shape}")
print(f"Clean Shape: {df_clean.shape}")
```

## 1. Catching Negative Revenue Anomaly
Revenue cannot logically be negative. In the pre-cleaned data, we caught 4 rows violating this boundary condition.

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df_raw['Revenue'], color='red', bins=30, kde=True)
plt.axvline(0, color='black', linestyle='--', label='Zero Boundary')
plt.title("Pre-Cleaning: Revenue (Shows Negatives)")
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(df_clean['Revenue'], color='green', bins=30, kde=True)
plt.axvline(0, color='black', linestyle='--', label='Zero Boundary')
plt.title("Post-Cleaning: Revenue (Imputed Positives)")
plt.legend()

plt.tight_layout()
plt.show()
```

## 2. Policy Support Score Scale Correction
The dataset natively used a 0-10 scale. Our backend spec mathematically expects the engineered scale to exist between integer boundaries 1-4.

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.countplot(data=df_raw, x='Policy_Support_Score', color='salmon')
plt.title("Pre-Cleaning: Policy Score (0-10 scale)")

plt.subplot(1, 2, 2)
sns.countplot(data=df_clean, x='Policy_Support_Score', color='lightgreen')
plt.title("Post-Cleaning: Policy Score (1-4 scale)")

plt.tight_layout()
plt.show()
```

## 3. Financial Risk vs Target Categories (Cleaned Data)

```python
df_clean['Is_Default'] = (df_clean['Financial_Risk_Level'] == 'High').astype(int)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.countplot(data=df_clean, x='Financial_Risk_Level', order=['Low', 'Medium', 'High'])
plt.title("Target Imbalance: Risk Level")

plt.subplot(1, 2, 2)
sns.boxplot(data=df_clean, x='Is_Default', y='Debt_to_Equity')
plt.title("Debt-to-Equity relative to Default")

plt.tight_layout()
plt.show()
```
